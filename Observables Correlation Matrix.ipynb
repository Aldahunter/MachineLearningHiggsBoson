```python
from dwrangling.pipelines \
    import get_dataframe
from dwrangling.dataframes \
    import ODataFrame
from dwrangling.dataframes.observables \
    import all_observables, add_reconstructed_observables

from analysis.plotting.observables \
    import plot_correlation_matrix_heatmap, plot_correlation_row
```

## Import Data 

```python
save_figures = True
set_colorbar_range = True

coll = "pp_2mu2e"
df = get_dataframe(coll)
```

```python
a_df = ODataFrame(add_reconstructed_observables(df))
s_df = a_df.get_s()
b_df = a_df.get_b()
```

```python
dfs = [a_df, s_df, b_df]
names = ['all_data', 'signal', 'background']
c_ranges = [(-1.00, 1.00), (-1.00, 1.00), (-1.00, 1.00)]

df_name_crange = list(zip(dfs, names, c_ranges))
```

# Signal Correlations

```python
fig_name = "signal_correlation_bar" if save_figures else None
fig = plot_correlation_row(a_df, collision=coll, fig_name=fig_name,
                           direction='v')
```

```python
fig_name = "abs_signal_correlation_bar" if save_figures else None
fig = plot_correlation_row(a_df, collision=coll, fig_name=fig_name,
                           direction='v', abs_corr=True)
```

# Observable Correlations

```python
fontsize = 18

corrs, observables = [], all_observables[:-1]

for df, name, c_range in df_name_crange:
    c_range = c_range if set_colorbar_range else (None, None)
    fig_name = "all_observables_" + name if save_figures else None
    fig, corr = plot_correlation_matrix_heatmap(df[observables],
                                                fig_name=fig_name,
                                                collision=coll,
                                                rtn_corr=True,
                                                c_range=c_range,
                                                alt_xticks=True,
                                                alt_yticks=False,
                                                fontsize=fontsize)
    
    corrs.append( corr )
```
