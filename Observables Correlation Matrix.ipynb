```python
from importlib import reload
import pandas as pd

import dwrangling.pipelines
from dwrangling.pipelines import get_dataframe
import dwrangling.dataframes
from dwrangling.dataframes import ObservablesDataFrame as ODF
import dwrangling.dataframes.observables
from dwrangling.dataframes.observables import add_reconstructed_observables
import analysis.plotting.observables
from analysis.plotting.observables import plot_correlation_matrix_heatmap, plot_correlation_row
```

## Import Data 

```python
save_figures = False
set_colorbar_range = True

coll = "pp_2mu2e"
df = get_dataframe(coll)
```

```python
a_df = ODF(add_reconstructed_observables(df))
s_df = a_df.get_s()
b_df = a_df.get_b()
```

```python
dfs = [a_df, a_df]#, s_df, b_df]
names = ['all_data', 'signal', 'background']
c_ranges = [(-0.56, 0.90),(-1.0, 1.0)]#, (-0.66, 1.00), (-0.45, 1.00)] #(-0.56, 0.90)
df_name_crange = list(zip(dfs, names, c_ranges))
```

# Signal Correlations

```python
fig_name = "signal_correlation_bar" if save_figures else None
signal_corr = plot_correlation_row(a_df, collision=coll, fig_name=fig_name)
```

# Observable Correlations

```python
corrs = []
for df, name, c_range in df_name_crange:
    c_range = c_range if set_colorbar_range else (None, None)
    fig_name = "all_observables_" + name if save_figures else None
    corrs.append( plot_correlation_matrix_heatmap(df[df.columns[:-1]], c_range=c_range,
                                                  collision=coll, fig_name=fig_name) )
```

# X-Observable Correlations

```python
corrs_x, observables = [], ['mu-_px', 'mu+_px', 'e-_px', 'e+_px']
for df, name, c_range in df_name_crange:
    c_range = c_range if set_colorbar_range else (None, None)
    fig_name = "x_observables_" + name if save_figures else None
    corrs_x.append( plot_correlation_matrix_heatmap(df[observables], c_range=c_range,
                                                    collision=coll, fig_name=fig_name) )
```

```python
corrs_y, observables = [], ['mu-_py', 'mu+_py', 'e-_py', 'e+_py']
for df, name, c_range in df_name_crange:
    c_range = c_range if set_colorbar_range else (None, None)
    fig_name = "y_observables_" + name if save_figures else None
    corrs_y.append( plot_correlation_matrix_heatmap(df[observables], c_range=c_range,
                                                    collision=coll, fig_name=fig_name) )
```
