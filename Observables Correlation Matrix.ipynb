```python
# Compute the correlation matrix
corr = all_df.corr()
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
def latex_label(observables):
    latex_observables = []
    for s in observables:
        ss = s.replace('mu', '\mu').replace('pT', 'p_{T}')
        if s.startswith('Z_'):
            ss = '$'+ss.replace('_','^{',1).replace('_','}_{',1)+'}$'
        elif s == 'signal':
            continue
        else:
            ss = ss.replace('-', '^{-}',1).replace('+', '^{+}',1)
            ss = '$'+ss.replace('_','_{',1)+'}$'
        latex_observables.append(ss)
    return latex_observables
```

```python
sns.set(style="white")

plot_all_df = all_df[n]
plot_all_df.columns = latex_label(list(plot_all_df.columns))# + ['signal']
plot_corr = plot_all_df.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(plot_corr, mask=mask, cmap=cmap, center=0, vmin=-0.5, #vmax=0.3,
            square=True, linewidths=.5, cbar_kws={"shrink": .75})
```

```python
def latex_observables(df):
    plot_observables = {}
    for s in df.columns:
        ss = s.replace('mu', '\mu').replace('pT', 'p_{T}')
        if s.startswith('Z_'):
            ss = '$'+ss.replace('_','^{',1).replace('_','}_{',1)+'}$'
        else:
            ss = ss.replace('-', '^{-}',1).replace('+', '^{+}',1)
            ss = '$'+ss.replace('_','_{',1)+'}$'
        plot_observables[s] = ss
    return df.rename(columns=plot_observables)
```

```python
sns.set(style="white")

plot_signal_correlations = signal_correlations.transpose(copy=True)
plot_signal_correlations = latex_observables(plot_signal_correlations)

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(plot_signal_correlations, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": 0.7, "orientation": "horizontal", 'pad':-0.15})
```
