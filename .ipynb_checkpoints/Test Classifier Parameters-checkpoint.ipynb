```python
sgn_df, bkg_df = load_event_DataFrames('pp_h_2mu2e')
sgn_df = add_complex_observables(sgn_df)
bkg_df = add_complex_observables(bkg_df)
```

```python
input_observables = ['m_H', 'Z_mu_rap', 'Z_e_rap', 'mu-_px', 'e-_py','signal']
sgn_data = sgn_df[input_observables]
bkg_data = bkg_df[input_observables]
data = pd.concat([sgn_data,bkg_data])

kwarg_variations_ranges = {'max_depth': (0, 20),
                           'min_samples_split': (0, 8000),
                           'min_samples_leaf': (0, 6000)}
n_varaitions = 2

# Create varriatons lists for each kwarg

kwarg_variations = OrderedDict()
shape = []
for kwarg in kwarg_variations_ranges:
    start, finish = kwarg_variations_ranges[kwarg]
    print(kwarg, ':', start, 'to', finish)
    kwarg_variations[kwarg] = np.linspace(start, finish, n_varaitions)
    shape.append(n_varaitions)


kwarg_grid = np.zeros(shape, dtype=dict)
sn_grid = np.zeros(shape, dtype=dict)
s_rmsq_grid = np.zeros(shape, dtype=dict)
b_rmsq_grid = np.zeros(shape, dtype=dict)



# sn_results = pd.DataFrame(list(range(1, n_tests+1)),
#                           columns=['Repeat'])
# rmsq_results_s = pd.DataFrame(list(range(1, n_tests+1)),
#                               columns=['Repeat'])
# rmsq_results_b = pd.DataFrame(list(range(1, n_tests+1)),
#                               columns=['Repeat'])

print( np.array(np.meshgrid(*kwarg_variations.values()))[:,:,:]  )
print([kwarg_var for kwarg_var in kwarg_variations])

# kwarg_variations = list(np.linspace(start, finish, n_varaitions))
```

```python
rgfnsdasdf

print("The kwargs being tested are:")
print(' ; '.join([str(kwarg) for kwarg in kwarg_variations]))
print()

splits = []
for kwarg in kwarg_variations:
    print(f"# KWARG: {kwarg}")
    print("  Test: ", end='')
    
    sn_tests = []
    rmsq_tests_s = []
    rmsq_tests_b = []
    for test in range(1,n_tests+1):
        print(f"{test} ", end='')
        
        tree = BinaryTreeClassifier()
        classifier_kwargs = {'max_depth': 10,                 # 8 {10}
                             'min_samples_split': kwarg,      # 2500 {2000, 300}
                             'min_samples_leaf': 750}        # 750 {100}
        
        _, sn_ratio, rmsq = SN_classifier(tree, data, classifier_kwargs, rseed=0)
        
        sn_tests.append(sn_ratio)
        rmsq_tests_s.append(rmsq['s'])
        rmsq_tests_b.append(rmsq['b'])
    
    print()
    
    splits.append(str(kwarg))
    sn_results[str(kwarg)] = sn_tests
    rmsq_results_s[str(kwarg)] = rmsq_tests_s
    rmsq_results_b[str(kwarg)] = rmsq_tests_b
    
    print()
```
