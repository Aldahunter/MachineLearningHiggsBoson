```python
###COPIED IN###
def gini_impurity(partition_samples, total_samples):
    G = 1.0
    for partition_sample in partition_samples:
        G -= (partition_sample / total_samples)**2
    return G
```

```python
from random import sample
def bootstrap(data,  f=0.8, n=None):
    """Select a random sample of data, with length n [or the fraction of \
    data, f]."""
    
    if (not n) and (not f):
        raise ValueError(f"Must enter either n ({n}) or f ({f})")
        
    if (not n) and f:
        if (f <= 0 or f > 1.0):
            raise ValueError(f"f must be between 0 and 1 (not {f})")
        n = f * len(data)
    
    elif (n < 1 or n > len(data)):
        raise ValueError(f"n must be between 1 and {len(data)} (not {n})")
        
    return sample(data, int(n))
```

```python
class RandomForestClassifier(object):
    
    def plant_forest(self, train_data, impurity_fn=gini_impurity,
                     n_trees=50, bootstrap_frac=0.7, show_growth=False,
                     **kwargs):
        self._n_trees = n_trees
        self._forest = []
        for n in range(n_trees):
            tree = BinaryTreeClassifier()
            tree.plant_tree(bootstrap(train_data, bootstrap_frac),
                            impurity_fn=impurity_fn, **kwargs)
            self._forest.append(tree)
            
            if show_growth:
                print(f"#{n+1:2d} trees planted out of {n_trees:d} \
                        ({100*(n+1)/n_trees:.2f}%)")
    train = plant_forest

    def classify(self, datum):
        P = 0.0
        for tree in forest.forest:
            P += tree.classify(datum)
        return P / forest.n_trees
    
    @property
    def forest(self): return self._forest
    @property
    def n_trees(self): return self._n_trees
```
