```python
from random import sample
def bootstrap(data,  f=0.8, n=None):
    if (not n) and (not f):
        raise ValueError(f"Must enter either n ({n}) or f ({f})")
        
    if (not n) and f:
        if (f <= 0 or f > 1.0):
            raise ValueError(f"f must be between 0 and 1 (not {f})")
        n = f * len(data)
    elif (n < 1 or n > len(data)):
        raise ValueError(f"n must be between 1 and {len(data)} (not {n})")
        
    return sample(data, int(n))
```

```python
def forest_classify(datum, forest):
    P = 0
    for tree in forest.forest:
        P += tree.classify(datum)
    return P / len(forest.forest)
```

```python
def test_classifier2(test_data, forest):
    results = []
    for datum in test_data:  # datum = (observable, label)
        p = forest_classify(datum, forest)
        results.append((*datum, p))
    return results
```
